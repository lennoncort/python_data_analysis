```python
import matplotlib.pyplot as plt
import numpy as np

class Circle(object):
    
    def __init__(self, radius = 3,color = "red"):
        self.radius = radius
        self.color = color
        
    def add_radius(self,r):
        self.radius = self.radius + r
    def change_color(self,c):
        self.color = c
    def drawCircle(self):
        fig, ax = plt.subplots()        
        circle = plt.Circle((0, 0), self.radius, color=self.color, fill=True)        
        ax.add_artist(circle)
        ax.set_xlim(-self.radius-1, self.radius+1)
        ax.set_ylim(-self.radius-1, self.radius+1)
        ax.set_aspect('equal', 'box')
        ax.set_title(f"Circle of radius {self.radius} and color {self.color}")
        plt.show()
        
C3 = Circle()
C3.add_radius(3)
C3.change_color("purple")
C3.drawCircle()

C4 = Circle()
C4.add_radius(10)
C4.change_color("red")
C4.drawCircle()
```

```python
## Hands on Lab
import matplotlib.pyplot as plt
##Create class
class Circle(object):
    ## Constructor
    def __init__(self, radius = 3, color = "red"):
        self.radius = radius
        self.color = color
    ##Method
    def add_radius(self, r):
        self.radius = self.radius + r
        return(self.radius)
    ##Method
    def change_color(self, c):
        self.color = c
        return(self.color)
    ##Method
    def drawCircle(self):
        plt.gca().add_patch(plt.Circle((0, 0), radius=self.radius, fc=self.color))
        plt.axis('scaled')
        plt.show()
```

```python
RedCircle = Circle(10,"red")
dir(RedCircle)
```

```python
RedCircle.add_radius(2)
RedCircle.radius
```

```python
RedCircle.color = "blue"
RedCircle.drawCircle()
```

```python
BlueCircle = Circle(radius = 100)
#dir(BlueCircle)
BlueCircle.radius
BlueCircle.color
BlueCircle.change_color("blue")
print('Color of object of radius 100 after applying the method change_color(c):',BlueCircle.color)
```

```python
##Rectangle Class
class Rectangle(object):

    def __init__(self, height, width, color):
        self.height = height
        self.width = width
        self.color = color

    def add_height(self, h):
        self.height = self.height + h
        return(self.height)

    def add_width(self, w):
        self.width = self.width + w
        return(self.width)

    def drawRectangle(self):
        plt.gca().add_patch(plt.Rectangle((0, 0), self.width, self.height ,fc=self.color))
        plt.axis('scaled')
        plt.show()

GreenRectangle = Rectangle(5, 5, "green")    
```

```python
GreenRectangle.add_width(5)
GreenRectangle.drawRectangle()
```
